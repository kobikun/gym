```python
raw_file="2015_annual_salary.txt.raw"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
import locale
locale.setlocale(locale.LC_NUMERIC,'')
locale.atof('3,000')
```

```python
df = pd.read_csv(raw_file, sep='\t',index_col=0,thousands=',')
```

```python
df.describe()
```

```python
df['population_ratio']=df['population']/df['population'].sum()*100
```

```python
df['total salary']=df['population']*df['avg_salary']
```

```python
df['total salary ratio']=df['total salary']/df['total salary'].sum()*100
```

```python
sns.kdeplot(df.population_ratio)
sns.kdeplot(df['total salary ratio'])
```

```python
?sns.distplot
```

```python

sns.distplot(df['total salary ratio'],hist=False,rug=True,color='y')
sns.distplot(df['population_ratio'],hist=False,rug=True,color='r')

```

```python
plt.hist(df['total salary ratio'],alpha=0.3)
sns.rugplot(df['total salary ratio'])
```

```python
pop_sal = sns.load_dataset(df)
sns.violinplot(x="dist",y="range", data=pop_sal)
```

```python
sns.heatmap([df['total salary ratio'],df['population_ratio']],annot=True)
```
